```python
even=0
odd=0
numberInput = input("Sample numbers(ex.1,2,3,...) = ")
numberSplit=numberInput.split(',')
for n in numberSplit:
    if int(n)%2==0:
        even+=1
    else:
        odd+=1
print("Number of even numbers :",even )
print("Number of odd numbers :",odd )
```
